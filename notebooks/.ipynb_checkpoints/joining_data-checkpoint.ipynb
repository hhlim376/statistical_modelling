```python
## Import stuff
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df_cb = pd.read_csv('stations.csv')
df_info = pd.read_csv('info.csv')
```

Join the data from Part 1 with the data from Part 2 to create a new dataframe.

```python
# rename column
df_cb = df_cb.rename(columns = {'id':'station_id'})
```

```python
## Merge dataframes from part 1 and 2
df_final = pd.merge(df_cb, df_info, on = 'station_id')
```

```python
## Drop columns we are not interested in 
df_final = df_final.drop(columns = ['lat', 'long'])
```

```python
## Check for and drop null values
df_final.isna().sum()
df_final = df_final.dropna()
```

```python
df = df_final.drop(columns = ['b_price', 'c_price'])
```

```python
import scipy
```

```python
df.to_csv('df_final.csv', index = False)
```

```python
## This is the final dataframe we want to work with, with extraneous details/information removed
df.describe()
```

Provide a visualization that you used as part of your EDA process. Explain the initial pattern or relationship you discoved through this visualization. 

From the scatter plot, there doesn't seem to be a very clear correlation between number of bikes at a bike station and any of our independant variables. Just visually, the closest positive correlation seems to be with the average bar rating and number of bars within a 500m radius of the location.

There does seem to be a pretty strong positive correlation between the number of bars and number of cafes at a location, but thats not what we are really looking at for this project

```python

sns.pairplot(data=df)
plt.show()
```

Using the heatmap, we see that the closest positive correlation for number of bikes is actually the number of cafes followed by the number of bars in a location. That being said it is still a very weak positive correlation overall.

As expected the strongest positive correlation is between number of cafes and number of bars but again, thats not what we are looking at for this project

```python
sns.heatmap(df.corr(), annot = True)
plt.show()
```

Looking at a boxplot, we can usually use it to detect outliers, in this case however, looking through the results in the dataframe as well as the boxplot I feel there are too many datapoint with high values to truly be considered outliers, so I won't be removing them from the dataset.

```python
sns.boxplot(data = df[['num_bikes', 'num_bars', 'num_cafes']], whis = 1.5)
plt.show()
```

# Database

Put all your results in an SQLite3 database (remember, SQLite stores its databases as files in your local machine - make sure to create your database in your project's data/ directory!)

```python
import sqlite3
from sqlite3 import Error
```

```python
## Create database
conn = sqlite3.connect('project.db')
```

```python
## Create cursor
c = conn.cursor()
```

```python
## Create table
c.execute("""CREATE TABLE data_final (
            station_id TEXT,
            num_bikes INTEGER,
            b_pop INTEGER,
            b_rating INTEGER,
            num_bars INTEGER,
            c_pop INTEGER,
            c_rating INTEGER,
            num_cafes INTEGER)
""")
```

```python
conn.commit()
```

```python
## Load data from dataframe into created table
df.to_sql('data_final',conn,if_exists='replace',index=False)
```

```python
conn.commit()
```
